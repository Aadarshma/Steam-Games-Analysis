```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('steam.csv')
df.head()
```

```python
df.tail()
```

```python
df['release_date']=pd.to_datetime(df['release_date'],errors='coerce')
df.info()
```

```python
df['release_date']=df['release_date'].dt.year
df.head()
```

```python
df['P.Rating_Ratio']=df['positive_ratings']/(df['positive_ratings']+df['negative_ratings'])
```

```python
df
```

```python
df.head()
```

```python
df2=df[['name','release_date','publisher','genres','owners','price','P.Rating_Ratio','negative_ratings']]
```

```python
df2
```

```python
df2.columns=['Name','Year','Publisher','Genres','Owners','Price','PositveRR','NegativeR']
df2
```

```python
df2['Genres']=df2['Genres'].str.split(';')
df2=df2.explode('Genres').reset_index(drop=True)
```

```python
df2.head()
```

```python
df2['Owners'] = df2['Owners'].astype('object')

df2.info()
df2
```

```python
df2['Ownerslow'] =df2['Owners'].str.split('-').str[0].astype(int)
df2['Ownershigh'] =df2['Owners'].str.split('-').str[1].astype(int)
df2['Avg_Owners']=(df2['Ownershigh']+df2['Ownerslow'])/2
df2=df2.drop(['Ownerslow','Ownershigh'],axis=1)
df2
```

```python
df5=df2[['Name', 'Year', 'Publisher', 'Genres', 'Avg_Owners', 'Price', 'PositveRR']]
```

```python
df5
```

```python
genre_count=df2['Genres'].value_counts().reset_index()
genre_count.columns=['Genres','genre_count']
genre_count.head()
```
